# Function Ardguments

```python
def func(*argv):
    print(type(argv))
    print(argv)

func(1,2,3)
```

```python
def func(**karg):
    print(type(karg))
    print(karg)

func(a=1,b=2)
```

```python
def func(*argv, **karg):
    print(type(argv))
    print(argv)
    print(type(karg))
    print(karg)

func(2,3,a=1,b=2)
```

```python
def func(a,b, *argv, g, **karg):
    print(type(argv))
    print(argv)
    print(type(karg))
    print(karg)
    print(g)
func(1,2,3,4,g=100,c=1,d=2)
```

```python
def f():
    y = 1
    def g(x):
        print(2*y+1)
        return y
    g(1)
    return g # return g_func
w = f()
print(type(w))
w(1)
```
