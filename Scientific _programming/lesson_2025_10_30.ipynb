```python
import numpy as np
```

```python
arr_uint8 = np.array([0b11101010,
                0b10101010,
                0b10101010,
                0b11111010,
                0b11001100,
                0b11001100,
                0b11001100,
                0b11111100, 
                ], dtype=np.uint8)
```

```python
print("Исходные uint8:")
for x in arr_uint8:
    print(f"{x:08b}")
```

```python
arr_uint32 = (arr_uint8).byteswap(False).view(np.uint32)
for x in arr_uint32:
    print(f"{x:32b}")
```
