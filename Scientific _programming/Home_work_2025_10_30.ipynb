```python
from math import sqrt
```

```python
def cross_out_value(array, value):
    for i in range(value * value, len(array), value):
        array[i] = False
```

```python
n = 100000000
```

```python
%%timeit
array = [True for x in range(0, n+1)]

end_len = int(sqrt(len(array))) + 1

for i in range(2, end_len):
    if array[i] == True:
        cross_out_value(array, i)
        #print(array)

counter = 0

for i in range(2, len(array)):
    if (array[i] == True):
        counter += 1
        # print(i, end="\t")

print(counter)
```

```python
import numpy as np
```

```python
def cross_out_value_numpy(n):
    is_prime = np.ones(n + 1, dtype=np.bool)
    is_prime[0] = False
    is_prime[1] = False
    for i in range(2, int(np.sqrt(n)) + 1):
        is_prime[i*i : n+1 : i] = False

    return np.where(is_prime==True)[0]
```

```python
%%timeit
cross_out_value_numpy(n)
```

```python
from numba import jit, prange
import numpy as np
```

```python
@jit(nopython=True)
def cross_out_value_numba(n):
    is_prime = np.ones(n + 1, dtype=np.bool)
    is_prime[0] = False
    is_prime[1] = False

    sqrt_n = int(np.sqrt(n))
    
    for i in prange(2, sqrt_n + 1):
        if is_prime[i]:
            for j in range(i * i, n + 1, i):
                is_prime[j] = False

    primes = []
    for i in range(2, n + 1):
        if is_prime[i]:
            primes.append(i)
    return np.array(primes)
```

```python
%%timeit
cross_out_value_numba(n)
```
